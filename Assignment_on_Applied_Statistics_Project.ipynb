## **Task 2 - Descriptive Statistics üìä**

````python
# Chart - 1 visualization code
# Calculate descriptive statistics for key variables
key_variables = ['Income', 'MntMeatProducts', 'MntWines', 'MntFruits', 'MntFishProducts', 'MntGoldProds']

# Create a DataFrame to hold the statistics
desc_stats = pd.DataFrame(index=key_variables)

# Measures of Central Tendency
desc_stats['Mean'] = ds1[key_variables].mean()
desc_stats['Median'] = ds1[key_variables].median()
desc_stats['Mode'] = ds1[key_variables].mode().iloc[0]  # Getting the first mode

# Measures of Dispersion
desc_stats['Variance'] = ds1[key_variables].var()
desc_stats['Standard Deviation'] = ds1[key_variables].std()

desc_stats
````

````python
# Function to identify and handle outliers
def handle_outliers(ds1, column):
    Q1 = ds1[column].quantile(0.25)
    Q3 = ds1[column].quantile(0.75)
    IQR = Q3 - Q1

    # Defining outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = ds1[(ds1[column] < lower_bound) | (ds1[column] > upper_bound)]
    print(f'Outliers in {column}:\n', outliers)

    # Handling outliers (you can choose to remove or cap them)
    # Here, we will replace them with NaN
    ds1[column] = ds1[column].mask((ds1[column] < lower_bound) | (ds1[column] > upper_bound), other=None)

# Identify and handle outliers for key variables
for var in key_variables:
    handle_outliers(ds1, var)

# Optionally, you can drop the rows with NaN values if you've removed outliers
# df = df.dropna()
````

## **Task 3 - Probability Distributions üé≤**

````python
# Chart - 2 visualization code
import numpy as np
from scipy.stats import binom

# Example: Binomial Distribution for offer acceptance
n = len(ds1)  # Total number of customers
p = ds1['Response'].mean()  # Probability of acceptance (mean response)

# Expected value for a binomial distribution
expected_value_binom = n * p

# Calculate probabilities (e.g., probability of exactly k successes)
k = 30  # For example, probability of exactly 30 customers accepting the offer
probability_k_successes = binom.pmf(k, n, p)

print(f'Expected Value (Binomial): {expected_value_binom}')
print(f'Probability of exactly {k} successes: {probability_k_successes}')
````

````python
from scipy.stats import norm, shapiro

# Example: Normal Distribution for Income
mean_income = ds1['Income'].mean()
std_income = ds1['Income'].std()

# Check for normality
stat, p_value = shapiro(ds1['Income'])
print(f'Shapiro-Wilk Test Statistic: {stat}, p-value: {p_value}')

# Calculate probabilities (e.g., probability of a customer earning less than a certain amount)
amount = 50000  # Example amount
probability_less_than = norm.cdf(amount, mean_income, std_income)

# Expected value for the normal distribution
expected_value_norm = mean_income  # In a normal distribution, the mean is also the expected value

print(f'Expected Value (Normal): {expected_value_norm}')
print(f'Probability of income less than {amount}: {probability_less_than}')
````

````python
from scipy.stats import poisson

# Example: Poisson Distribution for purchases
lambda_purchases = ds1['NumWebPurchases'].mean()  # Mean number of web purchases

# Expected value for Poisson distribution
expected_value_poisson = lambda_purchases

# Probability of observing k purchases
k_purchases = 5  # Example number of purchases
probability_k_purchases = poisson.pmf(k_purchases, lambda_purchases)

print(f'Expected Value (Poisson): {expected_value_poisson}')
print(f'Probability of observing {k_purchases} purchases: {probability_k_purchases}')
````
